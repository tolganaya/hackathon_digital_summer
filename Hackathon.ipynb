```python
import requests
from bs4 import BeautifulSoup

URL = "https://archive.sensor.community/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
stuff = soup.find_all('pre')
soup_pre = stuff[1]
soup_pre
```

```python
links_from_a = []
date_of_links = []
for a in soup_pre.find_all('a', href=True): 
    links_from_a.append('https://archive.sensor.community/'+a['href'])
    date_of_links.append(a['href'])
links_from_a
```

```python
fo = open("datesCorrect.txt", "r")
file_contents = fo.read()
parsed_links = []
for x in file_contents.split('\n'):
    for i in range(len(links_from_a)):
        if links_from_a[i][-11:-1]==x:
            parsed_links.append(links_from_a[i])
parsed_links
```

```python
csv_links =[]
for i in range(3,len(parsed_links)): #len(parsed_links)
    a_href_page = requests.get(parsed_links[i])
    a_href_soup = BeautifulSoup(a_href_page.content, "html.parser")
    a_href_soup2 = a_href_soup.find_all('a', href=True)
    for j in range(len(a_href_soup2)):
        if a_href_soup2[j].text[-4:] == '.csv'and (a_href_soup2[j].text[11:14]=='hpm' or a_href_soup2[j].text[11:18]=='pmsXXXX' or a_href_soup2[j].text[11:18]=='PPD42NS' or a_href_soup2[j].text[11:17]=='sds011' or a_href_soup2[j].text[11:16]=='sps30') and (a_href_soup2[j].text[-10:-3]=='_43254.' or a_href_soup2[j].text[-10:-3]=='_47621.' or a_href_soup2[j].text[-10:-3]=='_47623.'):
            csv_links.append(a_href_soup2[j].text)
csv_links
#print(csv_links)
```
